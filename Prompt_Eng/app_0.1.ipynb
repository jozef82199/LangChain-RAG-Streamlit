```python
# ✅ Create a prompt template with a variable
prompt = ChatPromptTemplate.from_template("calc {count}  pluse 2.")

# ✅ Create the chain
chain = LLMChain(llm=llm, prompt=prompt)

# ✅ Run the chain with required variable(s)
response = chain.run(count=2)

# ✅ Print the result
print(response)
```
