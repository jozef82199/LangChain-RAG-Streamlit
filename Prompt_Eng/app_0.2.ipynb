```python

from dotenv import dotenv_values
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain.schema import HumanMessage, SystemMessage, AIMessage

# Load API key from .env
env_values = dotenv_values("./app.env")
openai_api_key = env_values.get("OPENROUTER_API_KEY")


# Initialize the model , openai/gpt-4o-mini
llm = ChatOpenAI(
    model_name="openai/gpt-oss-20b:free",
    openai_api_base="https://openrouter.ai/api/v1",
    openai_api_key=openai_api_key,
)
```

```python
# âœ… Create a prompt template with a variable
prompt = ChatPromptTemplate.from_template("calc {count}  pluse 2.")

# âœ… Create the chain
chain = LLMChain(llm=llm, prompt=prompt, verbose=True)

# âœ… Run the chain with required variable(s)
response = chain.run(count=2)

# âœ… Print the result
print(response)
```

```python
from langchain.callbacks import get_openai_callback
with get_openai_callback() as cb:
    response = chain.run(count=2)
    
    # Print token statistics
    print(f"\n--- Token Usage ---")
    print(f"Prompt tokens: {cb.prompt_tokens}")
    print(f"Completion tokens: {cb.completion_tokens}")
    print(f"Total tokens: {cb.total_tokens}")
    print(f"Total cost (USD): ${cb.total_cost}")
    print(f"\nResponse: {response}")
```

```python
# Create the template
string_prompt = ChatPromptTemplate.from_template(
    'explain {subject} like i am 5. write this answer in {language}'
)

# Create a chain
chain = LLMChain(llm=llm, prompt=string_prompt)

# Run the chain
response = chain.run(subject='machine learning', language='Arabic')

print(response)
```

```python
base_conversation = [
    SystemMessage(content = 'you are a chief'),
    AIMessage(content = 'where are you from'),
    HumanMessage(content = 'Egypr'),
    AIMessage(content = 'Do you like spicy food?'),
    HumanMessage(content = 'not too much'),
]

ai_response = llm(base_conversation)

print(ai_response)
```

```python
base_conversation = [
    SystemMessage(content="You are a chief, you want to ask me some things to find out my favorite Asian dish"),
]

ai_response = None

for i in range(3):

    ai_response = llm( base_conversation )

    print("AI: ", ai_response.content)

    ## append the ai_response
    base_conversation.append( ai_response )

    user_response = HumanMessage(content=input("Your Answer:"))

    ## append the human_response
    base_conversation.append(user_response)

human_final_message = HumanMessage(content="And now. Which plate do you think I will like. Just type the plate name")
base_conversation.append(human_final_message)

final_ai_response = llm(base_conversation)
print( final_ai_response )
```
